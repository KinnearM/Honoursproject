```python
import extradim
```

```python
data='distpostGW19.txt' #Posterior samples marginalised over all params but luminosity distance.
z=0.438, 0.00003 #Redshift of EM signal source galaxy
distprior='Planck15' #Has to be the distance prior used for original GW parameter estimation. Read the associated release paper.
priors='Planck18' #Pick your priors! If you set this to 'flat' it breaks! Neat-o. I'll add a SH0ES option soon I promise. 
```

```python
example=extradim.leakageparam(data, z=z, distprior=distprior, priors=priors) #Initialise
```

```python
example.plotlikelihood(EM=True) #The distance posterior distribution has been converted to likelihood and renormalised. Let's have a look.
print(example.EMdist) #This is in Mpc
```

```python
sampleno=1000
nwalkers=100
#chosen to keep it under a minute, you probably want to do more of both
```

```python
example.sample(sampleno,nwalkers) #Sets up and runs emcee.
```

```python
example.ratio()
```
